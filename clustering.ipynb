```python
import scipy.stats as stats
    
def custom_barplot(df, col1=''):
    # based on https://github.com/Azure/Azure-TDSP-Utilities/blob/master/DataScienceUtilities/DataReport-Utils/Python/IDEAR-Python-AzureNotebooks.ipynb
    fig, axes = plt.subplots(2,2)
    axes = axes.reshape(-1)
    df[col1].plot(ax=axes[0], kind='hist')
    axes[0].set_title('Histogram of {}'.format(col1))
    df[col1].plot(ax=axes[1], kind='kde')
    axes[1].set_title('Density Plot of {}'.format(col1))
    ax3 = plt.subplot(223)
    stats.probplot(df[col1], plot=plt)
    axes[2].set_title('QQ Plot of {}'.format(col1))
    df[col1].plot(ax=axes[3], kind='box')
    axes[3].set_title('Box Plot of {}'.format(col1))
    
    fig.suptitle(f'GRAPHICS INFOS OF {col1} COLUMN', color='blue', fontsize=12)
    fig.tight_layout(pad=3.0)
    
for c in df.columns:
    custom_barplot(df, c)      
```

```python
def numerical_relations(df, col1, col2):
    x = df[col2]
    y = df[col1]
   
    return round(scipy.stats.pearsonr(x, y)[0], 6)

all_pearson = pd.DataFrame()
for c in list(df.columns):
    cols = []
    for c1 in list(df.columns):
        cols.append(numerical_relations(df, c, c1))
    all_pearson[c] = cols
    
all_pearson['input_col'] = list(df.columns)
all_pearson.set_index('input_col', inplace=True)
all_pearson
```

## Modeling

### Preprocessing data

```python
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv('./clusterization_data/clusterization.csv').drop(columns=['x3'])
scaler = MinMaxScaler()
x_values = scaler.fit_transform(df)
```

### Cluster groups analysis

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
cluster_to_search = list(range(2,10))
davies_bouldin_results = []
for n_clusters in cluster_to_search:
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)

    cluster_labels = clusterer.fit_predict(x_values)
    davies_bouldin_avg = davies_bouldin_score(x_values, cluster_labels)
    print("For n_clusters =", n_clusters,
          "davies_bouldin is :", davies_bouldin_avg)
    davies_bouldin_results.append(davies_bouldin_avg)
```

```python
f'davies_bouldin CLUSTER SIZE SUGGESTION: {cluster_to_search[np.argmin(davies_bouldin_results)]}, score: {np.min(davies_bouldin_results)}'
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=cluster_to_search, 
                              metric='calinski_harabasz', 
                              timings=False, locate_elbow=False)
print('calinski_harabasz Analysis')
visualizer.fit(x_values)
visualizer.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=cluster_to_search, metric='silhouette',
                              timings=False, locate_elbow=False)

print('silhouette Analysis')
visualizer.fit(x_values)
visualizer.show()  
```

# Suggested cluster size
- previous results analysis:
    - Based on silhouette and calinski harabasz the best cluster is 2, because that size can achieve the highest separation of the clusters;
    - Bases on davies_bouldin score the best cluster size is 9;
- Conclusion:
    - I use 3 scores to find the optimal cluster size, two of theses scores suggested the 2 as the best cluster size. In this way, I ***select the value 2 as the size of the cluster***
